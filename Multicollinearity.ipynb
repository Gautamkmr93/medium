```python
data=pd.read_csv('house_data.csv')
data.head()
```

```python
data=data.dropna()
data.head()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor 
  
# creating dummies for Swimming Pool 
data['Swimming Pool'] = data['Swimming Pool'].map({'Yes':1, 'No':0})
data['Location']=data['Location'].map({'Patna':0,'Kolkata':1,'Chennai':2,'lucknow':3,'New Delhi':4,'Mumbai':5,'Bangalore':6})  
# the independent variables set 
X = data[['Living_Space(sqft)','No_Of_Bedrooms','No_Of_Bathrooms','Condition(on scale of 5)','Year_Build','Swimming Pool','Location',]]   
# VIF dataframe 
vif_data = pd.DataFrame() 
vif_data["feature"] = X.columns   
# calculating VIF for each feature 
vif_data["VIF"] = [variance_inflation_factor(X.values, i) 
                          for i in range(len(X.columns))] 

print(vif_data)
#vif_data
```
