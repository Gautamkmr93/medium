```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('house_data.csv')
data.head()
```

```python
data.isna().any()
```

```python
Newdata=data.drop(['Swimming Pool','Location'],axis=1)
```

```python
Newdata.head()
```

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=10,metric='nan_euclidean')
newdata=imputer.fit_transform(Newdata)
newdata=pd.DataFrame(newdata)
newdata.head()
```

```python
from sklearn.impute import IterativeImputer
from sklearn.linear_model import BayesianRidge
imputer = IterativeImputer(BayesianRidge())
impute_data = pd.DataFrame(imputer.fit_transform(Newdata))
impute_data.head()
```

```python
data.fillna(method='ffill', inplace=True)
data.head()
```

```python
#fill NaN values using Zeros or using a fixed constant values
data['No_Of_Bedrooms']=data['No_Of_Bedrooms'].fillna(0)
data['No_Of_Bathrooms']=data['No_Of_Bathrooms'].fillna(100)
data['Living_Space(sqft)']=data['Living_Space(sqft)'].fillna(1000)
data['Condition(on scale of 5)']=data['Condition(on scale of 5)'].fillna(3)
data['Swimming Pool']=data['Swimming Pool'].fillna(data['Swimming Pool'].value_counts().index[0])
data['Location']=data['Location'].fillna(data['Location'].value_counts().index[0])
data['Price']=data['Price'].fillna(1000000)
data.head()
```

```python
data['Living_Space(sqft)']=data['Living_Space(sqft)'].fillna(data['Living_Space(sqft)'].mean())
data['No_Of_Bedrooms']=data['No_Of_Bedrooms'].fillna(data['No_Of_Bathrooms'].mean())
data['No_Of_Bathrooms']=data['No_Of_Bathrooms'].fillna(data['No_Of_Bathrooms'].mean())
data['Condition(on scale of 5)']=data['Condition(on scale of 5)'].fillna(data['Condition(on scale of 5)'].mean())
data['Swimming Pool']=data['Swimming Pool'].fillna(data['Swimming Pool'].value_counts().index[0])
data['Location']=data['Location'].fillna(data['Location'].value_counts().index[0])
data['Price']=data['Price'].fillna(data['Price'].mean())
data.head()
```
